```python
import numpy as np
import pandas as pd
import pyreadr

data_path = "/data/IDEA_DeFi_Research/Data/Lending_Protocols/Aave/V2/Mainnet/preprocessed_user_market_time/preloaded_4_None/checkpoints/aave_small_longSeq/checkpoint-169150/"
# Load embeddings
embeddings = np.load(f"{data_path}cls_embeddings.npy")
embedding_ids = np.load(f"{data_path}cls_embeddings_row_ids.npy")
data_df = pd.DataFrame(embeddings)
data_df["id"] = embedding_ids
# Load labels (assuming labels file is available)
labels_df = pyreadr.read_r(f"/data/IDEA_DeFi_Research/Data/Lending_Protocols/Aave/V2/Mainnet/id_types.rds")[None]
```

```python
x_ltm_path = "/data/IDEA_DeFi_Research/Data/Survival_Data/Borrow/X_ltm_seqLen50/"
```

```python
# Step 1: Merge the two dataframes on 'id'
merged_df = pd.merge(data_df, labels_df, on='id', how='inner')
```

```python
merged_df.head()
```

```python
merged_df.to_csv('~/merged_df_ltm_seqLen50.csv', index=False)
```

```python
# Step 2: Filter to just "borrow" transactions
borrow_df = merged_df[merged_df["type"] == "borrow"]

# Drop the 'type' column since it's no longer needed
borrow_embeddings_df = borrow_df.drop(columns=['type'])

# Inspect the result
print(borrow_embeddings_df.head())
```

```python
pyreadr.write_rds(f"{x_ltm_path}/X.rds", borrow_embeddings_df)
```
